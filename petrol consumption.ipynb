```python
x = data[['outlier_free_petrol_tax','Average income (dollars)','outlier_free_paved_highways','outlier_free_Population_Driver_licence']]
y = data['outlier_free_consumption_of_petrol']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
predictions = regressor.predict(x_test)
```

```python
predictions
```

```python
df = pd.DataFrame({'actual': y_test,'Predict': np.round(predictions,0)})
```

```python
from sklearn import metrics
```

```python
print('mean absolute error:',metrics.mean_absolute_error(y_test, predictions))
print('mean squared error:',metrics.mean_squared_error(y_test, predictions))
print('root mean squared error:',np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('r squared value:',metrics.r2_score(y_test,predictions))
```
